```python
# Importar librerías
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el dataset Iris
df = sns.load_dataset('iris')
df.head(5)
```

```python
print(df.info())
```

# Graficos

## a) Gráfico de barras:
Objetivo: Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.

Instrucciones: Utiliza un gráfico de barras para mostrar la media de las columnas petal_length y petal_width agrupadas por species.

```python
# Calcular promedio de petal_length y petal_width por especie
promedios = df.groupby('species')[['petal_length', 'petal_width']].mean()

# Crear gráfico de barras
promedios.plot(kind='bar', figsize=(8,5), color=['burlywood','mediumseagreen'])
plt.title('Promedio de Largo (lenght) y Ancho (widht) de Petalos para cada especie')
plt.xlabel('Especie')
plt.ylabel('Promedio (cm)')
plt.xticks(rotation=0)
plt.legend(title='Medida')
plt.show()
```

## b) Histograma:
Objetivo: Visualizar la distribución de los largos de los sépalos (sepal_length).

Instrucciones: Crea un histograma de la variable sepal_length y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

```python
plt.figure(figsize=(8,5))
plt.hist(df['sepal_length'], bins=20, color='steelblue', edgecolor='ivory')
plt.title('Distribución del Largo del Sépalo')
plt.xlabel('Largo del Sépalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
```

## c) Gráfico de dispersión (scatter plot):
Objetivo: Relacionar el largo del pétalo con el ancho del pétalo.

Instrucciones: Crea un gráfico de dispersión que compare petal_length con petal_width. Usa un color diferente para cada especie.

```python
plt.figure(figsize=(8,5))
colors = {'setosa':'red','versicolor':'blue','virginica':'green'}
for especie in df['species'].unique():
    subset = df[df['species'] == especie]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=especie, color=colors[especie])
plt.title('Relación del largo y ancho del petalo')
plt.xlabel('Petal_length (cm)')
plt.ylabel('Petal_width (cm)')
plt.legend()
plt.show()
```

## d) Gráfico de cajas (box plot):
Objetivo: Analizar la distribución de las medidas del sépalo (sepal_length, sepal_width).

Instrucciones: Crea un gráfico de cajas para mostrar la distribución de las variables sepal_length y sepal_width por especie.


```python
plt.figure(figsize=(10,6))
df.boxplot(column=['sepal_length','sepal_width'], by='species', notch=True, grid=True, patch_artist=True, vert=True)
plt.suptitle('Distribución de largo y ancho del sepalo por especie',fontsize=14)
plt.show()
```

## e) Gráfico de líneas:
Objetivo: Mostrar la tendencia de la longitud del pétalo (petal_length) a través de las observaciones.

Instrucciones: Crea un gráfico de líneas utilizando la variable petal_length (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
plt.figure(figsize=(8,5))
plt.plot(df.index, df['petal_length'], marker='o', linestyle='-', color='darkblue')
plt.title('Tendencia de la Longitud del Petalo a lo largo de cada observaciones')
plt.xlabel('Numero de renglon (observacion)')
plt.ylabel('Largo del Petalo (cm)')
plt.show()
```
