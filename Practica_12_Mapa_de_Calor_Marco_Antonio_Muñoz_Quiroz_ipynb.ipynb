```python
import seaborn as sns
import plotly.graph_objects as go

# 1. Cargar el dataset "flights"
flights = sns.load_dataset("flights")

# 2. Reorganizar los datos (años en filas, meses en columnas)
flights_pivot = flights.pivot(index="year", columns="month", values="passengers")
flights_pivot
```

```python
# 3. Crear el mapa de calor con plotly.graph_objects
fig = go.Figure(
    data=go.Heatmap(
        z=flights_pivot.values,             # Valores (número de pasajeros)
        x=flights_pivot.columns,            # Meses (columnas)
        y=flights_pivot.index,              # Años (filas)
        colorscale="Viridis",               # Escala de colores
        colorbar=dict(title="Pasajeros")    # Etiqueta de la barra de colores
    )
)

# 4. Personalizar el gráfico
fig.update_layout(
    title="Mapa de calor del número de pasajeros de vuelos (1949-1960)",
    xaxis_title="Mes",
    yaxis_title="Año",
    template="plotly_white"
)

fig.show()
```
